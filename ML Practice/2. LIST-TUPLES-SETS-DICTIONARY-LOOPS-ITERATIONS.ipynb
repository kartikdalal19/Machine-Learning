```python
#                                                                        SETS
```

```python
# SET - A set is an unordered, mutable collection of unique (no duplicate elements) elements.
# It is created using " {} "
# Indexing is not possible for sets as it is unordered collection of elements
# Thus slicing is not possible in sets
```

```python
# EMPTY

# empty set
a = set()
b = {}
# empty list
c = []
# empty tuple
d = ()

print(type(a),type(b),type(c),type(d))
```

```python
# remove() - to remove the element from the set

a = {1,2,3,4,5,'rahul','Rahul'}
print(a)
a.remove(3)
print(a)
```

```python
# update() - Updates the set
a = {1,2,3,4}
b = {5,6,7,8}
a.update(b)
print(a)
```

```python
update
b = {4,5,6,7,8}
print(a.union(b))
print(a|b)
```

```python
# intersect - To find the intersection of the sets
a = {1,2,3,4}
b = {3,4,5,6,7,8}
print(a.intersection(b))
print(a&b)
```

```python
# difference -
a = {1,2,3,4}
b = {3,4,5,6,7,8}
print(a-b)
```

```python
# symmetric_difference() - Returns elements that are in either set, but not in both.
a = {1,2,3,4}
b = {3,4,5,6,7,8}
print(a.symmetric_difference(b))
print(a^b)
```

```python
# Find all the prime numbers from 1 to 100
for i in range(2,101):
  is_prime = True
  for j in range(2,i):
    if i%j == 0:
      is_prime = False
      break;
  if is_prime == True:
    print(i)

```

```python
# Print the squares of only those numbers whuich are even

a = [1,2,3,4,5,6,7,8,9]
even_list = []
odd_list = []
for i in a:
  if i%2 == 0:
    even_list.append(i**2)
    # print(f'THE SQUARE OF {i} (EVEN) IS {i**2}')
  else :
    odd_list.append(i**3)
    # print(f'THE CUBE OF {i} (ODD) IS {i**3}')

print(f'EVEN LIST : {even_list} \nODD LIST : {odd_list}')
```

#### Chart - 3

````python
# # Chart - 3 visualization code

# # Do hosts with more listings charge more?


# Group by listing count and get average price
avg_price_host_count = df.groupby('calculated_host_listings_count')['price'].mean()

# Optionally limit to reasonable host counts to avoid long tail
avg_price_host_count = avg_price_host_count[avg_price_host_count.index <= 30]

# Plot
plt.figure(figsize=(12, 6))
avg_price_host_count.plot(kind='line', marker='o', color='blue')
plt.title('Average Price vs Host Listing Count')
plt.xlabel('Host Listing Count')

plt.ylabel('Average Price')
plt.grid(True)
plt.show()
````

##### 1. Why did you pick the specific chart?

***Line graph*** : Line graph is used to make the patterns visible.

##### 2. What is/are the insight(s) found from the chart?

* Casual Hosts Are More Competitive:
Individual hosts (with 15 listings) tend to keep prices lower to attract guests.

* Hosts with 10-11 listings have the hightest average price  .
